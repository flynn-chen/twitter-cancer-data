```python
# importing libraries and packages
import snscrape.modules.twitter as sntwitter
import pandas as pd
import datetime
import pytz
```

```python
# Creating list to append tweet data 
tweets_list1 = []
start_date = pytz.utc.localize(datetime.datetime.strptime("2020-08-01", '%Y-%m-%d'))

# Using TwitterSearchScraper to scrape data and append tweets to list
for i, tweet in enumerate(sntwitter.TwitterHashtagScraper('multiplemyeloma').get_items()): #declare a username 
    if tweet.date < start_date: #number of tweets you want to scrape
        break
    if i % 1000 == 0:
        print(i, tweet)
    tweets_list1.append([tweet.date, tweet.id, tweet.content, tweet.username]) #declare the attributes to be returned
    
# Creating a dataframe from the tweets list above
tweets_df1 = pd.DataFrame(tweets_list1, columns=['Datetime', 'Tweet Id', 'Text', 'Username'])
tweets_df1.to_csv("multiplemyeloma.2020-08-01.csv.gz")
tweets_df1
```
